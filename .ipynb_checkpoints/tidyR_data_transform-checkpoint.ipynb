#### Chapter 5: Data transform



* ```filter()```: Pick rows by their values.   
* ```arrange()```: Reorder the rows.  
* ```select()```: Pick variables by their names.  
* ```mutate()```: Create new variables (columns) with functions of existing variables and add to the end (right side).   
* ```summarise()```: Collapse many values down to a single summary.  
* SQL like operations:  
```min_rank() row_number() dense_rank() percent_rank() cume_dist() ntile()```

These can all be used in conjunction with ```group_by()```. 

1. The first argument is a data frame. 
2. The subsequent arguments describe what to do with the data frame, using the variable names (without quotes).
3. The result is a new data frame.

##### Tips:
* ```filter()```   
Computers use finite precision arithmetic (they obviously can’t store an infinite number of digits!) so remember that every number you see is an approximation. Instead of relying on```==```, ```use near()```.
Logical operators:```&``` is “and”, ```|``` is “or”, and ```!``` is “not”.

* ```select()```  
use ```rename()``` for rename varialbles
helpers: 
```starts_with("abc")```: matches names that begin with “abc”.  
```ends_with("xyz")```: matches names that end with “xyz”.  
```contains("ijk")```: matches names that contain “ijk”.  
```matches("(.)\\1")```: selects variables that match a regular expression. This one matches any variables that contain repeated characters. You’ll learn more about regular expressions in strings.  
```num_range("x", 1:3)```: matches x1, x2 and x3.  
```every_thing()```:```select(flights, time_hour, air_time, everything())``` moves time_hour, air_time to the front
* na.rm = T: used in aggregation functions such as ```mean()```, to remove missing values
* ```n()```: counts. ```sum(!is.na(x))``` for non-missing values. Good to keep track of counts so that values are based on large number of data but not small number of data. To count the number of distinct (unique) values, use ```n_distinct(x)```.
* skill or luck:  
    - http://varianceexplained.org/r/empirical_bayes_baseball/ 
    - http://www.evanmiller.org/how-not-to-sort-by-average-rating.html.
* aggregation functions:  
    -``` mean(x)```: mean  
    -```sd(x)```: standard deviation  
    -```IQR(x)```: interquantile range  
    -```mad(x)```: median absolute deviation  
    -```min(x), quantile(x, 0.25), max(x). ```   
    -```first(x), nth(x, 2), last(x)```   
* ```group_by``` undo with ```ungroup()```

### ```mutate()```

````r
flights_s <- select(flights, 
  year:day, 
  ends_with("delay"), 
  distance, 
  air_time
)
mutate(flights_s,
  gain = dep_delay - arr_delay,
  speed = distance / air_time * 60
) %>% head(3)
````

### ```summarise()```

````r
summarise(flights, delay = mean(dep_delay, na.rm = TRUE))
````

````r
by_day <- group_by(flights, year, month, day)
by_day%>%head(3)
````

### Pipe

````r
delays <- flights %>% 
  group_by(dest) %>% 
  summarise(
    count = n(),
    dist = mean(distance, na.rm = TRUE),
    delay = mean(arr_delay, na.rm = TRUE)
  ) %>% 
  filter(count > 20, dest != "HNL")%>% 
    ggplot(mapping = aes(x = dist, y = delay)) +
      geom_point(aes(size = count), alpha = 1/3) +
      geom_smooth(se = FALSE)
````

````r
delays
````

### counts

````r
# Convert to a tibble so it prints nicely
batting <- as_tibble(Lahman::Batting)

batters <- batting %>% 
  group_by(playerID) %>% 
  summarise(
    ba = sum(H, na.rm = TRUE) / sum(AB, na.rm = TRUE),
    ab = sum(AB, na.rm = TRUE)
  )

batters %>% 
  filter(ab > 100) %>% 
  ggplot(mapping = aes(x = ab, y = ba)) +
    geom_point(aes( size = 1), alpha = 0.1) + 
    geom_smooth(se = FALSE)
#> `geom_smooth()` using method = 'gam' and formula 'y ~ s(x, bs = "cs")'
````

### ```group_by()``` in conjunction with ```mutate()``` ```filter()``` and ```summarize()```

````r
# Find the most delayed each group:

flights_s %>% 
  group_by(year, month, day) %>%
  filter(rank(desc(arr_delay)) < 10) %>%
    head(3)
````

````r
# Find all groups bigger than a threshold:

popular_dests <- flights %>% 
  group_by(dest) %>% 
  filter(n() > 365)
popular_dests %>% head(3)
````

````r
# Standardise to compute per group metrics
popular_dests %>% 
  filter(arr_delay > 0) %>% 
  mutate(prop_delay = arr_delay / sum(arr_delay)) %>% 
  select(year:day, dest, arr_delay, prop_delay) %>%
    head(3)
````
